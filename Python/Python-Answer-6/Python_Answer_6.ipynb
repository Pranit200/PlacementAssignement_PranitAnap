# Attributes 
id: Identification Number

num: Number of the Pokémon in the official Pokédex

name: Pokémon name

img: URL to an image of this Pokémon

type: Pokémon type

height: Pokémon height

weight: Pokémon weight

candy: type of candy used to evolve Pokémon or given when transfered

candy_count: amount of candies required to evolve

egg: Number of kilometers to travel to hatch the egg

spawn_chance: Percentage of spawn chance (NEW)

avg_spawns: Number of this pokemon on 10.000 spawns (NEW)

spawn_time: Spawns most active at the time on this field. Spawn times are the same for all time zones and are expressed in local time. (NEW)

multipliers: Multipler of Combat Power (CP) for calculating the CP after evolution See below

weakness: Types of Pokémon this Pokémon is weak to

next_evolution: Number and Name of successive evolutions of Pokémon

prev_evolution: Number and Name of previous evolutions of Pokémon

```python
import pandas as pd
data = pd.read_excel('/content/df.xlsx')
data
```

```python
#Get all Pokemons whose spawn rate is less than 5%
import matplotlib.pyplot as plt
spawned_pokemon = data[data['spawn_chance']<0.05].sort_values('spawn_chance',ascending=True)
plt.figure(figsize=(10,10))
plt.barh(spawned_pokemon['name'],spawned_pokemon['spawn_chance'])
plt.title('Pokemons whose spawn rate is less than 5%')
plt.xlabel('spawn_chance')
plt.ylabel('Pokemons Name')
plt.show()
spawned_pokemon.head(2)
```

```python
#Get all Pokemons that have less than 4 weaknesses
weak_pokemon=data[data['weaknesses'].apply(lambda x: len(x.split(','))) < 4]
# Plotting the number of weaknesses for each Pokemon
weak_pokemon['num_weaknesses'] = weak_pokemon['weaknesses'].apply(lambda x: len(x.split(',')))
sorted_weak_pokemon = weak_pokemon.sort_values('num_weaknesses')

plt.figure(figsize=(20,20))
plt.barh(sorted_weak_pokemon['name'],sorted_weak_pokemon['num_weaknesses'], edgecolor='black')
plt.xlabel('Number of Weaknesses')
plt.ylabel('Pokemons Name')
plt.title('Pokemons that have less than 4 weaknesses')
plt.show()
sorted_weak_pokemon.head(2)
```

```python
#Get all Pokemons that have no multipliers at all
no_multiplier_pokemon = data[data['multipliers'].isna()]
# Calculate the number of Pokémon with no multipliers and with multipliers
Pokemon_no_multiplier = len(no_multiplier_pokemon)
Pokemon_with_multiplier = len(data) - Pokemon_no_multiplier

# Create a pie chart
labels = ['without multipliers', 'with multipliers']
sizes = [Pokemon_no_multiplier, Pokemon_with_multiplier]
colors = ['#FF9999', '#66B2FF']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pokémon')
plt.show()
no_multiplier_pokemon.head(2)
```

```python
#Get all Pokemons that do not have more than 2 evolutions
evol_pokemon = data[data['next_evolution'].apply(lambda x: len(eval(x))<=2 if pd.notnull(x) else False)]
# Plotting the number of evolutions for each Pokemon
evol_pokemon['num_evolution'] = evol_pokemon['next_evolution'].apply(lambda x: len(eval(x)))
sorted_evol_pokemon = evol_pokemon.sort_values('num_evolution')

plt.figure(figsize=(20,20))
plt.barh(sorted_evol_pokemon['name'],sorted_evol_pokemon['num_evolution'], edgecolor='black')
plt.xlabel('Number of Evolutions')
plt.ylabel('Pokemons Name')
plt.title('Pokemons that do not have more than 2 evolutions')
plt.show()
sorted_weak_pokemon.head(2)
```

```python
# Get all Pokemons whose spawn time is less than 300 seconds
# Defining function to convert time to second.
def time_to_seconds(time):
    try:
        minutes, seconds = str(time).split(":")
        minutes = int(minutes)
        seconds = int(seconds)
        total_seconds = minutes * 60 + seconds
        return total_seconds
    except ValueError:
        return None

data['spawn_time'] = data['spawn_time'].apply(time_to_seconds)
```

```python
time_lessthan_300 = data[data['spawn_time'] < 300].sort_values('spawn_time')

plt.figure(figsize=(20,20))
plt.barh(time_lessthan_300['name'],time_lessthan_300['spawn_time'], edgecolor='black')
plt.xlabel('Spawn Time')
plt.ylabel('Pokemons Name')
plt.title('Pokemons whose spawn time is less than 300 seconds')
plt.show()
time_lessthan_300.head(2)
```
