#Importing data

```python
import pandas as pd
df=pd.read_csv('/content/df.csv')
df.head()
```

#Get all the Earth meteorites that fell before the year 2000

```python
import datetime
# Convert the 'year' column to datetime format
df['year'] = pd.to_datetime(df['year'],errors='coerce')

# Extract the year
df['year'] = df['year'].dt.year
sorted_data = df[df['year'] < 2000].sort_values(ascending=False, by = 'year')
#Earth meteorites that fell before the year 2000
sorted_data[sorted_data['fall']=='Fell'].head(2)
```

#Get all the earth meteorites co-ordinates who fell before the year 1970

```python
required_data = df[(df['fall'] == 'Fell') & (df['year'] < 1970)]
# Coordinates of meterorites before 1970
coor_df = required_data[['reclat', 'reclong']]
coor_df
```
