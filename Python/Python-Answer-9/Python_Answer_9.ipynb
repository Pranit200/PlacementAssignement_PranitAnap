#Importing data

```python
import pandas as pd
df=pd.read_csv("https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD")
df.head(2)
```


‚óè Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
import matplotlib.pyplot as plt
cars_not_qualify = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility']=='Not eligible due to low battery range']
# Get the count of car types
car_type_counts = cars_not_qualify['Model'].value_counts()

# Plot a bar chart
plt.figure(figsize=(10, 6))
car_type_counts.plot(kind='bar')
plt.title('Car Types that Do Not Qualify')
plt.xlabel('Car Model')
plt.ylabel('Count')
plt.show()
```

‚óè Get all TESLA cars with the model year, and model type made in Bothell City.

```python
tesla_cars_in_bothell_city = df[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell')][['Model Year', 'Model']]
tesla_cars_in_bothell_city
```

```python
tesla_counts = tesla_cars_in_bothell_city['Model Year'].value_counts().sort_index()
# Create the bar chart
plt.bar(tesla_counts.index, tesla_counts.values)
plt.xlabel('Model Year')
plt.ylabel('Total Tesla Cars')
plt.title('Number of Tesla Cars Made in Bothell City')
plt.xticks(rotation=45)

plt.show()
```

‚óè Get all the cars that have an electric range of more than 100, and were made after 2015

```python
electric_cars = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)]
electric_cars
```

```python
# Create the graph
plt.figure(figsize=(10, 6))
plt.scatter(electric_cars['Model Year'],electric_cars['Electric Range'])
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.title('Range of Elect Year')

plt.show()
```

‚óè Draw plots to show the distribution between city and electric vehicle type

```python
import seaborn as sns
electric_cars_df=df[['City', 'Electric Vehicle Type']]
# Count the occurrences of each combination
count_df = electric_cars_df.groupby(['City', 'Electric Vehicle Type']).size().reset_index(name='Count')
top_cities = count_df.groupby('City')['Count'].sum().nlargest(15).index
count_df = count_df[count_df['City'].isin(top_cities)].sort_values(ascending=False,by= 'Count')
```

```python
import seaborn as sns

# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Count', hue='Electric Vehicle Type', data=count_df)
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution of Electric Vehicle Types by City')
plt.xticks(rotation=45)
plt.legend(title='Electric Vehicle Type')
plt.show()
```
