```python
import requests
import pandas as pd

def download_and_convert_to_csv(link):
    #Following line downloads the data from the link
    response = requests.get(link)
    df = response.json()
    
    # This line convert the data into a structured dataframe
    df = pd.DataFrame(df)
    
    # This line converts the dataframe to CSV format
    df.to_csv("df.csv", index=False)
    #Following line conforms once data is converted to csv.
    print("Data downloaded and converted to structured CSV format as df.csv")
    return df
# Provide the URL of the data 
url = "https://data.nasa.gov/resource/y77d-th95.json"

# Call the function to download and convert the data to CSV
download_and_convert_to_csv(url)
```
